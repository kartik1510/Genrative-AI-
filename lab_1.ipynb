```python
import torch
from diffusers import StableDiffusionPipeline
import os
import json


model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

prompts = [
    "An astronaut sitting in a cafe on Mars, realistic style",
    "A medieval castle floating in the clouds at sunrise",
    "A smart city with autonomous vehicles and green buildings",
    "A wildlife photographer capturing a tiger in a dense jungle",
    "An underwater city with glowing lights and futuristic architecture"
]

output_dir = "generated_images"
os.makedirs(output_dir, exist_ok=True)

prompt_image_list = []

for i, prompt in enumerate(prompts):
    print(f"Generating image {i+1}")

    image = pipe(prompt).images[0]

    image_name = f"image_{i+1}.png"
    image.save(os.path.join(output_dir, image_name))

    prompt_image_list.append({
        "prompt": prompt,
        "image": image_name
    })

with open(os.path.join(output_dir, "dataset_info.json"), "w") as f:
    json.dump(prompt_image_list, f, indent=4)

print("Done. Images generated using Hugging Face diffusion model.")
```
